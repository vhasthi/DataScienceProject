# Abstract

Recognizing the relationship between language and culture or how language can reveal much of a person’s thoughts, values, and beliefs, we want to know can we predict if an individual is classified as “Thinking (T)” or “Feeling (F)” in line with their Myers-Briggs Type Indicator (MBTI), based on the sentiment analysis of their social media posts to the PersonalityCafe forum? That is, we are motivated to explore users’ word choices and the sentiment value (positive, neutral, or negative) of their words in their posts from the PersonalityCafe forum to predict, if possible with high accuracy, whether that person contains a T or F in their MBTI type. We expect feelers to portray a more positive sentiment value than thinkers and use words such as “feel” or “love”. Contrastingly, we expect thinkers to be more objective and utilize more neutral or negative words than feelers such as “think” and “know”. We perform sentiment analysis using the VADER sentiment analyzer and the TF-IDF model. Our results show that both thinkers and feelers use the words ‘think’ and ‘people’ most often, which contradicts our hypothesis that thinkers and feelers would have words of different sentiment value as their most common words. However, feelers do use more words such as ‘love’ and ‘feel’ in their posts compared to thinkers which does support our hypothesis.

To predict thinking and feeling classification, we used logistic regression, kNN, decision trees, and SVM. They each performed with 53-54% accuracy which is not a high prediction score for Thinking/Feeling types based on sentiment analysis from online posts. These findings give rise to the conclusion that it is not possible to accurately predict Thinking/Feeling personality traits based on text analysis alone. In other words, there are many confounding factors not taken into consideration such as context, frequency of posts, other interactions on an online platform such as likes and comments, behavior outside the online environment, and thought patterns inconsistent with their posts or how they portray themselves to be online. Next steps would be to measure these influencing variables with a more comprehensive lexicon in order to improve the accuracy of our MBTI personality prediction models. Even though it is not possible to accurately predict a person’s MBTI Thinking or Feeling trait solely based on the sentiment analysis of their posts on the PersonalityCafe forum, this is a great stepping stone to achieve a higher understanding of the complexity and intricacy of the human personality as per MBTI.

## Background and Prior Work

According to the Myers Briggs official website, the goal of the Myers-Briggs Type Indicator (MBTI) test is to help people understand how they take in information, make decisions, and follow a pattern of either perception or judgment.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) There are 16 MBTI personality types that are made up of four preference pairs that showcase different aspects of personality, which are: 

Extraversion (E) or Introversion (I) 

Sensing (S) or Intuition (N)

Thinking (T) or Feeling (F)

Judging (J) or Perceiving (P)

The Myers-Briggs theory argues that one trait from each of the four preference pairs is innate to a person, meaning people are born with a specific set of personality preferences that then develop over time. However, life experiences and the environment can impact how one expresses their personality preferences, so they are not fixed, rather these traits are how a person tends to act or how their natural instinct dictates their behavior. Thus, in this project, we will see if a specific behavior, measured as the sentiment of the text someone posts online, is potentially dictated by aspects of their MBTI, specifically whether they are more Thinking (T) or Feeling (F). 

Whereas the introversion-extroversion scale describes people’s energy and the sensing-intuition scale describes people’s mind, the third aspect of MBTI, thinking vs feeling, describes people’s nature.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) To put it simply, this preference type refers to whether one values emotion (F) or rationality (T) more when making decisions. For instance, feelers (F) tend to follow their heart, value empathy and compassion, and have a high concern for others. On the other hand, thinkers (T) value objective knowledge and use logic and fairness when dealing with others. In terms of what someone posts online, this could translate to feelers posting more about themselves and their emotions, while thinkers would share more objective facts about topics they’re interested in, as opposed to sharing about themselves. Thus, a sentiment analysis would reveal feelers have a larger range of emotional content that they post, both positive and negative, while thinkers use more neutral language since they may tend to describe things more objectively. 

We will analyze the sentiment of online posts using the VADER lexicon from nltk<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3). This lexicon is especially useful for handling online posts since it takes into consideration several factors prevalent in social media text. For example, it accounts for words in all caps, repeated punctuation (e.g. ‘!!!’ or ‘???’), idioms (e.g. ‘cut the mustard’), and emoticons, among other factors. It then assigns a score between sentiment score to the sentence (in our project, the post) as a whole based on the tokenized elements comprising the sentence (post). This sentiment score contains four elements: a positive, negative, neutral, and compound score. The positive, negative, and neutral scores all range from 0 to 1, and the compound score essentially averages these all together to obtain a composite score from -1 to 1, where -1 is completely negative, 0 is completely neutral, and 1 is completely positive. For our project, we will perform sentiment analysis on each post in the dataset, and then use the compound sentiment score to try and predict whether that post came from a thinker or a feeler. For this binary classification task, we will implement Logistic Regression, KNN, Decision Trees, and SVM models, and then compare the results to determine which model had the highest accuracy in correctly predicting thinking/feeling.

We believe that utilizing sentiment scores to predict thinking/feeling classification of online posts is a novel approach because existing research using this dataset does not employ any sentiment analysis, rather it relies on classification models based off of the text itself. Previous research by Rajshree Varma<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4) implemented Random Forests models, XG Boost classifiers, and Logistic Regression, among several other models, in order to predict specific MBTI type of the post. They began by cleaning the posts to remove links, punctuation, repeated words, etc., and then they predicted the MBTI type based on these cleaned posts. However, none of these models got higher than 58% accuracy in predicting MBTI type; using a 60-40 train-test split, Logistic Regression and XDG Boost performed the best with 58% and 57% accuracy, respectively, and KNN classification performed the worst with only 16% accuracy. We hope that by implementing sentiment analysis and predicting MBTI based on sentiment score, we will be able to improve upon these results and achieve higher accuracy because the sentiment score accounts for some of the context of the posts, whereas the models in this research did not take into account any contextual elements. Additionally, we aim to classify only thinking versus feeling, as opposed to determining the exact MBTI personality type as this research did, so we hope that our models will perform better since binary classification is a simpler task.

### Section 3 of EDA - TF-IDF

In this section of EDA, we want to perform tf-idf on the posts to further see which words are the most important for thinkers and feelers. 

First, we want to remove some custom stopwords in addition to the default ‘english’ stopwords, since we found that these appear too often in the posts and don't contribute much meaning.

```python
# To start, set custom stopwords as the default English stopwords
custom_stopwords = set(stopwords.words('english'))

# Update the list of common stopwords to remove filler words and mbti types
custom_stopwords.update(['like', 'im', 'ive', 'esfj', 'enfp', 'entj', 'estj', 'intj', 'estp', 'isfj', 'istj'])

# Convert the set of custom stopwords to a list
custom_stopwords_list = list(custom_stopwords)

# Use the TfidfVectorizer with the custom stopwords list
tfidf = TfidfVectorizer(sublinear_tf=True,
                        analyzer='word',
                        max_features=2000,
                        tokenizer=word_tokenize,
                        stop_words=custom_stopwords_list)
```

We also want to remove punctuation from our text so the TF-IDF model only considers relevant *words* and not punctuation.

```python
def remove_punctuation(text):
    # make posts lowercase
    text = text.lower()

    #remove punctuation
    text = re.sub(r'[^a-zA-Z\s]', ' ', text)
    return text
```

We apply the remove function to our posts and get the most relevant words from posts for each MBTI type, organizing our data into 16 rows for each personality type. We then fit the TF-IDF model to our text, now having a representations of each personality type’s posts as numerical values. After this, we convert the tfidf_matrix into a dataframe using MBTI personality types as indices. Lastly, we get the most common word for each MBTI type by getting the word with the highest TF-IDF score.

```python
# Make a copy of the data frame to perform tf-idf on
df_tfidf = df2.copy()

# Remove punctuation from posts 
df_tfidf['posts'] = df_tfidf['posts'].apply(remove_punctuation)

# Group posts by their personality type
grouped_posts = df_tfidf.groupby('type')['posts'].apply(' '.join).reset_index()

# Extract 'posts' and 'type' columns
text = grouped_posts['posts']
mbti_type = grouped_posts['type']

# Fit and tranform post data using TF-IDF
tfidf_matrix = tfidf.fit_transform(text)

# Create dataframe with index as personality types and columns as most common word for each type
post_tfidf = pd.DataFrame(tfidf_matrix.toarray(), columns = tfidf.get_feature_names_out(), index = mbti_type)

# Find most common word for each personality type
most_common_word = post_tfidf.idxmax(axis=1)
print(most_common_word)
```

We convert the list of the most common words for each mbti into a dataframe.

```python
# Turn into a dataframe
df_most_common_word = pd.DataFrame(most_common_word)

# Rename column to "Most Common Word"
df_most_common_word = df_most_common_word.rename(columns={0: "Most Common Word"})
df_most_common_word
```

We initialize a new dataframe with MBTI type index from post_tfidf as our new index and we name the columns each as the first, second, and third top words for each MBTI type. We then identify the top 3 most common words used by each mbti type by iterating through each row and selecting the words with the top 3 largest TF-IDF scores.

# Sentiment Analysis

Now that we've compared the words thinkers and feelers use most frequently in their posts, it's time to finally analyze the sentiment of each post. Ultimately we will use these sentiment scores to see if we can predict thinking/feeling classification.

First, upon examining our current dataframe, the ‘posts’ column in each row contains 50 posts separated by ‘|||’. To effectively generate sentiment scores for individual posts, we initiate a process of splitting these concatenated strings within the ‘posts’ column. 

```python
# Define a function to split the 50 posts in each row
def split_posts(row):
    posts = row['posts'].split('|||')
    return pd.DataFrame({'type': [row['type']] * len(posts), 'posts': posts})

# Apply the function to each row and concatenate the results
new_df = pd.concat(df.apply(split_posts, axis=1).tolist(), ignore_index=True)
```

```python
new_df
```

Examining the updated dataframe, we noticed unnecessary web links that might not contribute to accurate sentiment analysis. To focus on the actual post content, we created a function called ‘remove_links’. This function uses a simple method to get rid of these links from each post in the ‘posts’ column, helping us concentrate on the essential content for our analysis.

```python
# Define a function to remove the links from posts
def remove_links(text):
    pattern = r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\\(\\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+'
    return re.sub(pattern, '', text)

# Apply the function to each row and concatenate the results
new_df = pd.concat(df.apply(split_posts, axis=1).tolist(), ignore_index=True)

# Remove links from 'posts' column in the new DataFrame
new_df['posts'] = new_df['posts'].apply(remove_links)
```

Now that our data is cleaner, we proceed with sentiment analysis on each post in the updated dataframe using the VADER sentiment analyzer. To accomplish this, we’ve developed a function called ‘get_sentiment_scores’, which takes an individual post as input and utilizes the VADER sentiment analyzer to calculate a single compound sentiment score. This score ranges from -1 to 1, representing the overall sentiment of the text. 

After defining this function, we apply it to the ‘posts’ column within the ‘new_df’ we previously created. Following this analysis, we perform a cleanup step by removing any empty rows from the ‘posts’ column, and their associated sentiment scores are retained for further analysis, having previously eliminated rows containing links and other unnecessary elements. 


```python
# Initialize the VADER sentiment analyzer
sid = SentimentIntensityAnalyzer()

# Function to perform sentiment analysis and return sentiment scores
def get_sentiment_score(post):
    return sid.polarity_scores(post)['compound']

# Apply the sentiment analysis function to the 'posts' column in the new DataFrame and create a new column for sentiment scores
new_df['Sentiment Score'] = new_df['posts'].apply(get_sentiment_score)

# Remove rows with empty 'posts'
new_df = new_df[(new_df['posts'] != '')]

new_df
```

We notice that there are still some rows that contain posts that appear empty or don't contain any actual words. Since these will not contribute accurately towards our sentiment analysis, we need to further remove these rows. We define a function to find the rows that don't contain words and remove them from the data frame.

```python
# Function to check if a post contains any words
def contains_words(post):
    tokens = word_tokenize(post)
    return any(token.isalpha() for token in tokens)

# Create a boolean mask for rows without words in the 'posts' column
mask_no_words = ~new_df['posts'].apply(lambda x: contains_words(x))

# Get the rows without words
rows_no_words = new_df[mask_no_words]
```

## Mann-Whitney U Test

We would like to perform a t-test to determine if there is a statistically significant difference in mean sentiment score between thinkers and feelers. The results of this test will either support or deny our hypothesis that thinkers have more neutral sentiment scores, while feelers have more positive sentiment scores. First, we must check that we meet the assumptions to perform a t-test. Let's take a look at the distribution of sentiment scores for thinkers and feelers and ensure that they are normally distributed.

```python
# Define thinkers and feelers
thinkers = df_filtered[df_filtered['Thinking/Feeling'] == 1]
feelers = df_filtered[df_filtered['Thinking/Feeling'] == 0]

# Plot histogram for Thinkers
plt.figure(figsize=(6, 4))
plt.hist(thinkers['Sentiment Score'], bins=20, color='blue', alpha=0.7, label='Thinkers')
plt.title('Distribution of Sentiment Scores for Thinkers')
plt.xlabel('Sentiment Scores')
plt.ylabel('Frequency')
plt.legend()
plt.show()

# Plot histogram for Feelers
plt.figure(figsize=(6, 4))
plt.hist(feelers['Sentiment Score'], bins=20, color='red', alpha=0.7, label='Feelers')
plt.title('Distribution of Sentiment Scores for Feelers')
plt.xlabel('Sentiment Scores')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

Uh oh, it looks like these distributions are not normal. Since we don't meet the normality assumption, we cannot perform a t-test, so we turn to the Mann-Whitney U test. Since this is a nonparametric test, it doesn't make any assumptions about the underlying distributions the two groups come from. Instead, this test assesses whethere there is a difference in distribution between two groups. While the results of this test cannot directly support or deny our hypothesis, since they won't indicate exactly *what* the sentiment score distributions of thinkers and feelers are, this test can at least tell us if the sentiment scores of posts from thinkers and feelers follow *different* distributions, which is still useful insight.

```python
# Perform Mann-Whitney U Test
stat, p_value = mannwhitneyu(feelers['Sentiment Score'], thinkers['Sentiment Score'])

# Take a look at test statistic and p-value
print('Test Statistic:', stat)
print('p-value:', p_value)
```

The analysis comparing sentiment scores of ‘thinkers’ and ‘feelers’ showed a big difference. The very low p-value (0.0) suggests this difference isn’t likely random chance, instead, it strongly suggests that thinkers and feelers come from two different distributions. We cannot explicitly conclude that thinkers tend to have more neutral posts, while feelers have more positive posts, but we have evidence that they at least have a different distribution of sentiments across their posts.

Just out of curiosity, let's take a look at the distributions of sentiment score for thinkers and feelers using a boxplot. This will provide a visualization of where the majority of data lies for the two classes.

```python
# Create box plot of sentiment score distributions for thinkers and feelers
sns.boxplot(x='Thinking/Feeling', y='Sentiment Score', data=df_filtered, hue_order=['Feeling', 'Thinking'])
plt.xlabel('Thinking/Feeling Classification')
plt.ylabel('Sentiment Analysis Scores')
plt.title('Distribution of Sentiment Analysis Scores for Thinkers and Feelers')
plt.xticks([0, 1], ['Feeling', 'Thinking'])
plt.show()
```

Based on this plot, it looks like the sentiment score of posts for thinkers is more neutral, while for feelers, it's more positive. This supports our hypothesis that thinkers use more neutral language in their posts while feelers use more positive language. However, since we cannot perform a t-test to measure if there is a statistically significant difference in sentiment scores between these two groups, we do not have rigorous evidence that our hypothesis is correct; the most we can say is that the two groups have different distributions of sentiment score, and it *looks* like thinkers are more neutral while feelers are more positive, but this is just speculation.

## Model Fitting

Now that we've seen that the sentiment scores of thinkers and feelers follow different distributions, we would like to see if we can predict thinking/feeling classification based off of sentiment score. We will implement 5 different models and compare their accuracies at classifying thinking/feeling. If all the models have high accuracy, this means sentiment score of online posts is a good predictor for determining thinking/feeling classification.

To begin, we need to define our predictor variable (X) and target labels (y). This will be the 'Sentiment Score' and 'Thinking/Feeling' columns of our dataframe, respectively. 

```python
# Make 'Sentiment Score' our predictor, and 'Thinking/Feeling' the label
mbti_X = df_filtered[['Sentiment Score']].values.reshape(-1, 1)
mbti_y = df_filtered[['Thinking/Feeling']].values.ravel()
```

Now we use an 80-20 train-test split so that the models are trained on 80% of the data, and then we predict on 20%. We set the random state to 200 so that all models we use will have the same training and test data so we can accurately compare their performances.

```python
# Define train and test sets based on 80-20 split
mbti_X_train, mbti_X_test, mbti_y_train, mbti_y_test = train_test_split(mbti_X, mbti_y, test_size=0.2, random_state=200)
```

### Logistic Regression

The first model we will implement is logistic regression. Logistic regression is a probabilistic model that outputs probabilities of belonging to a particular class and is trained to minimize the difference between predicted probabilities and actual class labels. We chose to use this model since it works well for simple binary classification tasks.

```python
# Implement logistic regression
LR_model = LogisticRegression()
LR_model.fit(mbti_X_train, mbti_y_train)
LR_y_pred = LR_model.predict(mbti_X_test)
LR_accuracy = accuracy_score(mbti_y_test, LR_y_pred)
LR_conf_mat = confusion_matrix(mbti_y_test, LR_y_pred)
LR_disp = ConfusionMatrixDisplay(LR_conf_mat)
LR_disp.plot()

# Look at accuracy and confusion matrix
print('Results (Logistic Regression):')
print("Accuracy:", LR_accuracy)
plt.show()
```

We see that the logistic regression model performs with 53.5% accuracy at correctly classifying thinking/feeling based on the sentiment score of the post. The confusion matrix shows us that the model does a lot better at predicting 0 (feeling), and tends to over-predict this classification.

### KNN

The next model we chose to implement is k-Nearest Neighbors. kNN is a simple algorithm that classifies a new data point based on the majority class of its *k* nearest neighbors in the training dataset. It's a non-parametric and instance-based algorithm, meaning it doesn't make strong assumptions about the underlying data distribution, which is good for our purposes since we already saw that sentiment scores don't follow a normal distribution. We chose to set the k-hyperparameter as 565 since this is roughly the square root of the number of datapoints in our training set, which is common practice in choosing the k-hyperparameter. 

```python
# Implement KNN 
KNN_model = KNeighborsClassifier(n_neighbors=565)
KNN_model.fit(mbti_X_train, mbti_y_train)
KNN_y_pred = KNN_model.predict(mbti_X_test)
KNN_accuracy = accuracy_score(mbti_y_test, KNN_y_pred)
KNN_conf_mat = confusion_matrix(mbti_y_test, KNN_y_pred)
KNN_disp = ConfusionMatrixDisplay(KNN_conf_mat)
KNN_disp.plot()

# Look at accuracy and confusion matrix
print('Results (KNN):')
print("Accuracy:", KNN_accuracy)
plt.show()
```

We see that the kNN model performs slightly better than logistic regression since the accuracy is 54.2%, compared to 53.5% for logisitic regression. However, this still isn't great. We also see that again, the model tends to over-predict the 0-label.

### Decision Trees

The next model we chose to implement is decision trees. Since we only have one predictor variable in our data (Sentiment Score), this model is very simple. It essentially chooses a threshold value, and then if the sentiment score is above this threshold, it predicts 'thinking' classification, while if the sentiment score is below this threshold, it predicts 'feeling'. This model is useful because of its simplicity and interpretibility, but the tradeoff is that it may seem pretty trivial.

```python
# Implement decision trees
tree_model = DecisionTreeClassifier()
tree_model.fit(mbti_X_train, mbti_y_train)
tree_y_pred = tree_model.predict(mbti_X_test)
tree_accuracy = accuracy_score(mbti_y_test, tree_y_pred)
tree_conf_mat = confusion_matrix(mbti_y_test, tree_y_pred)
tree_disp = ConfusionMatrixDisplay(tree_conf_mat)
tree_disp.plot()

# Look at accuracy and confusion matrix
print('Results (Decision Trees):')
print("Accuracy:", tree_accuracy)
plt.show()
```

We see that this model gets around the same accuracy of kNN and logistic regression since it performs with 53.9% accuracy. # Ethics & Privacy

We recognize there are limitations to classifying individuals as “Thinking (T)” or “Feeling (F) per the Myer-Briggs Type Indicator (MBTI) on a social media platform where body language, gestures, tone of voice, real life conversational context, and personal issues are not accounted for. We acknowledge individuals might portray an emotional state in their posts contrary to their emotional state in real life. Consequently, we are strictly classifying an individual as either “Thinking (T)” or “Feeling (F)” based on the emotional valence (positive, negative, neutral) of their word choice in their posts and do not consider alternative variables. 


Additional Concerns:
1. Privacy: Collecting data from individuals’ social media posts poses a privacy risk for identity reveal. Therefore, we make sure to use non-identifiable data (no usernames or sensitive information provided).

2. Bias and Stereotyping: using sentiment analysis to predict personality traits might bring up the discussion of stereotyping a person based on their word choices in the context of their posts. We recognize word choices do not fully portray a person’s personality and we acknowledge we use the standards laid out by the WKWSCI Sentiment Lexicon.

3. Ethics using Data Usage: We ensure to carefully analyze personal data by including a representative sample in our dataset (not leaving out certain individuals based on their posts), following analytical procedures, and not discarding data that goes against our hypothesis.

Instead, we had to turn to the non-parametric Mann-Whitney U test to perform statistical analysis between the sentiment scores of online posts for thinkers and feelers. This test revealed that the distribution of sentiment scores for thinkers and feelers follow different distributions, which somewhat supports our hypothesis. If we performed hyperparameter optimization using validation sets, or if we used a different, more robust lexicon than VADER, we might be able to slightly improve the performance of our models. However, we still likely wouldn’t get higher than around 60% accuracy, which means it’s not a problem with our models, it’s a problem with the data. Namely, sentiment score is not an accurate representation of the content and context of these posts, so it cannot accurately predict thinking/feeling classification. 

A personality encompasses various complex aspects of an individual’s character. So, predicting if a person is of type ‘Thinking’ and ‘Feeling’ solely off their social media posts may not fully encapsulate the multifaceted nature of personality traits. While social media posts can reflect some aspects of communication and expression, they may not truly capture the true thinking or feeling characteristic of an individual. It is also important to take into consideration that social media is a platform people use to present themselves however they wish to be perceived by peers, families and fellow users. The self representation can create a skewed representation of their personality traits, making it challenging to rely on social media data for an accurate assessment of their personalities. Individuals can tailor their content according to societal expectations, privacy concerns or even anonymity while creating a completely different identity on the web. 

For future research in this field of MBTI types and their relationship to social media posts, it would be helpful to include more information than just sentiment score when determining if you can accurately predict MBTI type from social media posts. For instance, if a pre-trained large language model were used to predict MBTI classification based on the context and content of social media posts, it would likely perform better than our models since LLMs already understand language, so more context and linguistic patterns of the posts would be considered than when using sentiment score. However, more than likely there are simply too many confounding variables to accurately predict MBTI type based on online posts, so perhaps considering posting patterns over an extended period of time, or the frequency of posts, or engagement patterns (likes and comments on other posts, behavior across multiple social media platforms, etc.) would be helpful.
