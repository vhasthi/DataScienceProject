# Research Question

Can we predict if an individual is classified as “Thinking (T)” or “Feeling (F)” in line with their Myers-Briggs Type Indicator (MBTI), based on the sentiment analysis of their social media posts to the PersonalityCafe forum?

## Background and Prior Work

According to the Myers Briggs official website, the goal of the Myers-Briggs Type Indicator (MBTI) test is to help people understand how they take in information, make decisions, and follow a pattern of either perception or judgment.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) There are 16 MBTI personality types that are made up of four preference pairs that showcase different aspects of personality, which are: 

Extraversion (E) or Introversion (I) 

Sensing (S) or Intuition (N)

Thinking (T) or Feeling (F)

Judging (J) or Perceiving (P)

The Myers-Briggs theory argues that one trait from each of the four preference pairs is innate to a person, meaning people are born with a specific set of personality preferences that then develop over time. However, life experiences and the environment can impact how one expresses their personality preferences, so they are not fixed, rather these traits are how a person tends to act or how their natural instinct dictates their behavior. Thus, in this project, we will see if a specific behavior, measured as the sentiment of the text someone posts online, is potentially dictated by aspects of their MBTI, specifically whether they are more Thinking (T) or Feeling (F). 

Whereas the introversion-extroversion scale describes people’s energy and the sensing-intuition scale describes people’s mind, the third aspect of MBTI, thinking vs feeling, describes people’s nature.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) To put it simply, this preference type refers to whether one values emotion (F) or rationality (T) more when making decisions. For instance, feelers (F) tend to follow their heart, value empathy and compassion, and have a high concern for others. On the other hand, thinkers (T) value objective knowledge and use logic and fairness when dealing with others. In terms of what someone posts online, this could translate to feelers posting more about themselves and their emotions, while thinkers would share more objective facts about topics they’re interested in, as opposed to sharing about themselves. Thus, a sentiment analysis would reveal feelers have a larger range of emotional content that they post, both positive and negative, while thinkers use more neutral language since they may tend to describe things more objectively. 

We will analyze the sentiment of online posts using two methods. For the first method, in order to predict if an individual is classified as “Thinking (T)” or “Feeling (F)” in line with their Myers-Briggs Type Indicator (MBTI), we will be using a reference lexicon to determine the emotional content of an individual's posts and to find the correlation between our prediction of positive, negative, or neutral content with their Myer Briggs personality type. Upon reviewing other research approaches of sentiment categorization and sentiment strength prediction using and comparing 6 different lexicons, “the WKWSCI lexicon obtained the best accuracy of 69%, which is substantially better than the other lexicons”<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) in sentiment categorization and demonstrated a “correlation between human scoring and WKWSCI was a moderate 0.57, which is significantly better (α = 0.01) than the correlation with SO-CAL lexicon”<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) in sentiment strength prediction. This showcases the effectiveness of using a lexicon for extracting the emotional content of text data and of assigning an emotional intensity value to it, such as analyzing the overall sentiment of 410 news headlines as outlined in the lexicon based sentiment analysis research paper. However, although the WKWSCI (Wee Kim Wee School of Communication and Information) lexicon does seem to be a potentially useful tool, one must be cognizant of its limitations and of supporting techniques that can improve its accuracy. For instance, one must look out for common sense knowledge and reasoning needed to determine sentiment and indirect or implicit expressions of satisfaction or dissatisfaction that the WKWSCI lexicon model lacks to adequately label or even recognize. It also exhibits conflicting polarities between different lexicons, coding different sentiments depending on the different contexts of the words in each lexicon. Fortunately, we can supplement the WKWSCI lexicon with, “a bag-of-words machine-learning model using SVM -which- obtained an accuracy of about 84% for document-level sentiment categorisation”<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3). All in all, lexicon based sentiment analysis with WKWSCI as a guiding tool will be a useful method to answer our research question, demonstrating the highest accuracy in sentiment categorization and a better performance in sentiment strength prediction among 5 other different lexicons. WKWSCI lexicon document-level performance can be improved with SVM (supporting vector machine) technique and WKWSCI lexicon on its own is great in “sentence-level and aspect-based sentiment analysis”.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) 

The second method we will use to measure the sentiment of a person’s online postings will be to employ the Hugging Face DistilBERT language model.<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4) The DistilBERT model, introduced by Sanh et al. in 2020, represents a distilled version of the large BERT (Bidirectional Encoder Representations from Transformers) model. The primary aim of DistilBERT is to provide a faster and more resource-efficient alternative to BERT while still retaining similar performance in various natural language processing tasks. DistilBERT achieves this by employing a knowledge distillation approach, where it is trained on a combination of labeled data and the output probabilities of a larger pre-trained model (in this case, BERT). The distilled model captures the essential contextual information within a text while significantly reducing the number of parameters.

In the context of our research project, DistilBERT will be utilized to analyze and understand the textual content of online posts from the PersonalityCafe forum. By fine-tuning the model on our MBTI and online postings dataset, we will use the model to predict whether an individual is thinking or feeling based on the presence of "T" or "F" in their MBTI type. Specifically, the training data for this model will be each individual post, which has an associated score of 0 or 1 for T or F, respectively. Then the model will predict the label. The reason we want to use this specific model is because it is already pre-trained, meaning it already has an understanding of language, and more importantly, it is distilled down to a smaller size, allowing us to run it on our own machines. We will alter the output values of the model to predict “T” or “F” instead of a specific sentiment in order to adapt the model to our specific task. 

The limitations of using DistilBERT are that this model is not interpretable, meaning that we cannot understand why it would predict “T” or “F”. This means that any hypothesis we come up with about how MBTI can be predicted from online postings cannot actually be tested since we cannot prove if it is correct. However, this model is useful because it can actually understand language, whereas our first method of sentiment analysis does not; the first method only considers counts of words and emotional content of specific words, but it does not understand anything about the structure or context of sentences as a whole, while DistilBERT does. Thus, DistilBERT can answer our question of whether we can predict if an individual is classified as “T” or “F” by looking at the meaning, context, and structure of the posts, it just can’t tell us why it makes the prediction it does. 

We believe that utilizing DistilBERT is a novel approach to determining thinking-feeling of online postings because existing research using this dataset does not employ any language models, rather it relies on more simple classification models. Previous research by Rajshree Varma<a name="cite_ref-5"></a>[<sup>5</sup>](#cite_note-5) implemented Random Forests models, XG Boost classifiers, and Logistic Regression, among several other models, in order to predict specific MBTI type of the post. However, none of these models got higher than 58% accuracy in predicting MBTI type; using a 60-40 train-test split, Logistic Regression and XDG Boost performed the best with 58% and 57% accuracy, respectively, and KNN classification performed the worst with only 16% accuracy. We hope that by implementing DistilBERT, we will be able to improve upon these results and achieve higher accuracy because DistilBERT understands contextual elements of language, whereas the models in this research did not. Additionally, we aim to classify only thinking versus feeling, as opposed to determining the exact MBTI personality type, which is a harder task.

Now that we've seen the number of posts per personality type, we want to see if there is any actual difference in these posts. First, we can determine if there is a difference in length of post for each personality type by plotting the average number of words per post for each personality type. We can visualize this with a boxplot to easily see the different distributions for each personality type. But first, we must define a function to count the average number of words per post for each row of the data. Then we can apply this function to our dataframe, and ultimately create a plot with this new column.

```python
# Define a function to count the average words per post
def average_words_per_post(row):
    # Split the posts using '|||'
    posts = row.split('|||')
    
    # Calculate the number of words for each post
    word_counts = [len(post.split()) for post in posts]
    
    # Calculate the average number of words per post
    average_words = np.mean(word_counts)
    
    return average_words
```

```python
# Apply this function to our dataframe and create a new column containing average words per post
df2['average_words_per_post'] = df2['posts'].apply(average_words_per_post)
df.head()
```

```python
# Plot a boxplot of the average number of words per post for each personality type
plt.figure(figsize=(12, 8))
sns.boxplot(x='type', y='average_words_per_post', data=df, palette='husl')
```

Based on this graph, it appears there is not too much of a difference in average length of post across the different personality types. However, there do appear to be many outliers for each personality type, which is interesting. Also, the variance for each personality type appears to be pretty large.

Now that we've visualized the average number of words per post for the different personality types, it's time to see if there is a difference in the actual content of the posts. We will visualize the most commonly used words among each personality type by using a WordCloud, where larger words are the more commonly used word for that personality type. This will help us see any major differences between personality type in what kinds of words they choose to use most often.

```python
# Define a function that generates a wordcloud for each personality type
def generate_wordcloud_for_type(df, mbti_type, ax):
    # Filter data for the given MBTI type
    type_data = df[df['type'] == mbti_type]

    # Combine all posts for the specific MBTI type into a single string
    text = ' '.join(type_data['posts'])

    # Generate word cloud 
    wordcloud = WordCloud(width=200, height=100, background_color='white').generate(text)

    # Plot the word cloud on the specified axis
    ax.imshow(wordcloud, interpolation='bilinear')
    ax.set_title(f'Word Cloud for {mbti_type} Type')
    ax.axis('off')
```

```python
# 16 unique MBTI types 
types = df['type'].unique()

# Create a 4x4 grid of subplots
fig, axes = plt.subplots(4, 4, figsize=(16, 16))
axes = axes.flatten()

# Loop through each MBTI type and generate the corresponding word cloud
for i, mbti_type in enumerate(types):
    generate_wordcloud_for_type(df, mbti_type, axes[i])

plt.tight_layout()
plt.show()
```
